```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use("ggplot")
```

```python
model = ["Standard_R50-LPIPS", "Salman2020Do_50_2-LPIPS", "Standard_R50-SSIM", "Salman2020Do_50_2-SSIM",]
DDPM = [0.32907,0.35255, 0.51848,0.56461]
GAATTACK = [0.32165,0.34340, 0.66072,0.59515]

# 创建分组柱状图，需要自己控制x轴坐标
xticks = np.arange(len(model))

fig, ax = plt.subplots(figsize=(10, 7))
# 所有门店第一种产品的销量，注意控制柱子的宽度，这里选择0.25
ax.bar(xticks, DDPM, width=0.25, label="AdvDDPM", color="red")
# 所有门店第二种产品的销量，通过微调x轴坐标来调整新增柱子的位置
ax.bar(xticks + 0.25, GAATTACK, width=0.25, label="GA-Attack", color="blue")

# ax.set_title("Grouped Bar plot", fontsize=15)
ax.set_xlabel("Model-index")
ax.set_ylabel("similarity")
ax.legend()

# 最后调整x轴标签的位置
ax.set_xticks(xticks + 0.12)
ax.set_xticklabels(model)
```
